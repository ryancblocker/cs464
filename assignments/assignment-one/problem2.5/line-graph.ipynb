```python
# Import pandas and matplotlib libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the csv file into a pandas dataframe
df = pd.read_csv("error%.csv")

# Create a new column for the trial number
df["Trial"] = df.index + 1

# Plot the line graph with x-axis as Trial and y-axis as Mouse and Trackpad
plt.plot(df["Trial"], df["Mouse"], label="Mouse")
plt.plot(df["Trial"], df["Trackpad"], label="Trackpad")

# Add labels, title, and legend to the graph
plt.xlabel("Trial")
plt.ylabel("Percent Error (%)")
plt.title("Mouse vs Trackpad Percent Error by Trial")

# Change the x-axis tick frequency to increments of 10
plt.xticks(ticks=range(0, 61, 10))

# Show the graph
plt.legend()
plt.show()
```
